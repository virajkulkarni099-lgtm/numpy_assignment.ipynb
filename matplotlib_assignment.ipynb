```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.figure(figsize=(4, 2))
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.figure(figsize=(5,3))
plt.bar(students, marks)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Highlight the region with highest revenue
explode = [0.1, 0, 0, 0]

plt.figure(figsize=(5, 4))
plt.pie(
    revenue,
    labels=regions,
    autopct='%1.1f%%',
    explode=explode,
    startangle=90
)
plt.title("Revenue Distribution by Region")
plt.axis('equal')  # Ensures proper circle
plt.show()
```
