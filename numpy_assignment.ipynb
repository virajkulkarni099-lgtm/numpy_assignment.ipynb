```python
import numpy as np

#1D array
array=np.arange(1,21)
array
```

```python
# To find the sum
sum=np.sum(array)
print(sum)
```

```python
#To find the mean
mean=np.mean(array)
print(mean)
```

```python
#To find the median
median=np.median(array)
print(median)
```

```python
#To find the standard deviation
deviation=np.std(array)
print(deviation)
```

```python
#Indices where the number is grater than the 10
np.where(array>10)
```

```python
# For the 4*4 dimension
array2=np.arange(1,17).reshape(4,4)
array2
```

```python
# To find the transpose
transpose=array2.T
print(transpose)
```

```python
# Row wise sum
sum1=np.sum(array2,axis=1)
print(sum1)
```

```python
# Column wise sum
sum2=np.sum(array2,axis=0)
print(sum2)
```

```python
# Random 3*3 
array3=np.random.randint(1,18,(3,3))
array4=np.random.randint(1,18,(3,3))
print("First array")
print(array3)
print("Second array")
print(array4)
```

```python
# Random operations on the array3 and array4
print(array3+array4)
print(" ")
print(array3-array4)
print(" ")
print(array3*array4)
```

```python
# Dot product
print(np.dot(array3,array4))
```
